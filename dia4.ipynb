challenge 6
Definir una clase madre que herede a sus clases hijas dos atributos de instancia. Mostrar ambos atributos en pantalla desde los objetos de las clases hijas.

Observación: Editar las clases hijas para que no se inicialicen.

```python
#Definir la clase madre usando de ejemplo materiales de papeleria mas comunes
class Materiales_academicos: 
    tipo='papeleria'

    def __init__ (self, tamanho, cantidad_hojas):
        self.tamanho=tamanho
        self.cantidad_hojas=cantidad_hojas

#Crear clases acuerdo a diversas carreras   
class Arquitectura(Materiales_academicos):
    trabajos= 'planos'
    def tipo_hojas (self):
        print (f'Para los {self.trabajos} se necesitan hojas de tamanho {self.tamanho}')


class HistoriayGeo(Arquitectura):
    trabajos= 'investigacion'
    def cuadernos(self):
        print (f'Para realizar trabajos de {self.trabajos} los estudiantes necesitan cuadernos de {self.cantidad_hojas} hojas.')

carrera1=Arquitectura('A0', '100')
carrera1.tipo_hojas()

carrera2=HistoriayGeo('A4', '200')
carrera2.cuadernos()
```

Crear dos personajes de una historia que se representen a traves de clases. Ejemplo: Un hereo y un villano.
Cada personaje debe tener poderes o acciones diferentes. Y deben presentarse en una frase.

Ejemplo: Yo me llamo pepito y tengo super fuerza
Ejemplo: Yo me llamo juanito y puedo volar

```python
#Crear la clase madre
class Personajes_comic:
    #Definir atributo de clase, en este caso es tipo de personaje. 
    personaje='superpoderoso'
    #Definir los atributos de instancia.
    def __init__ (self, superpoder, apodo):
        self.superpoder=superpoder
        self.apodo=apodo

#En la clase hija llamada Propositos pongo lo que serian los posibles propositos de los personajes
    
class Propositos(Personajes_comic):
    tarea1= 'ayudar a salvar el mundo.'
    tarea2= 'conqusitar la ciudad'
    def super_poder (self):
        print (f'Me llamo {self.apodo}. Mi poder es {self.superpoder}. Mi proposito es {self.tarea1}')

class Supervillano(Propositos):
    color= 'azul y negro'
    def color_traje (self):
        print (f'Soy {self.apodo}. Me representa el traje color {self.color} y tengo {self.superpoder}. Pronto voy a {self.tarea2}.')


superheroe1=Propositos('super velocidad', 'Flash')
superheroe1.super_poder()

supervillano1=Supervillano('super inteligencia', 'Cerebro')
supervillano1.color_traje()
    
    
```
