Cuando el usuario ingrese la contrasena correcta, imprimir un mensaje secreto.

```python
# Paso 1_ Definir la variable
contrasenha_oficial = "molly123"
# Paso 2_ Solicitar al usuario la contrasenha

while True: 
    contrasenha=input ("Ingrese la contrasenha: ") 
    if (contrasenha != contrasenha_oficial ):
        print ("Contraseña incorrecta! Ingrese nuevamente: ")

    else:
        print ("USUARIO CORRECTO. BIENVENIDA :)") #Mensaje secreto a imprimir
        break

        
    

```

challenge 4
Crear una lista de 5 elementos. Luego, seguir los siguientes pasos.

Agregar un elemento al final de la lista.
Modificar el segundo elemento.
Eliminar el tercer elemento.

```python
colores=["azul", "naranja", "verde", "rojo", "lila"]
print (f"Lista original: ", colores)

colores.append("amarillo") #Agregar un elemento al final de la lista con el comando append
print (f"Lista modificada 1: ", colores)

colores[1] = "celeste" #Modificar el elemento de la 2da posicion
print (f"Lista modificada 2: ", colores)

colores.pop(2) #Eliminar elemento con el comando POP
print (f"Lista modificada 3: ", colores)
```

challenge 5
Escribir un programa que itere en un rango de 10 veces. Por cada iteracion multiplicar el iterador por dos e imprimir el resultado.

```python
# Definir nombre de mi iterador ---- iterador

for iterador in range(10):
    # Realizar operacion de multiplicacion
    multiplicacion= iterador*2  
    # Imprimir el resultado. Usar el "+1" para mostrar que son 10 multiplicaciones
    print (f"La multiplicacion {iterador+1} da: ", multiplicacion)
```

challenge 6
Recorrer los elementos de una lista de marcas e imprimir cada marca solamente si tiene 6 letras o mas

```python
lista_de_marcas = ["samsung", "nike", "xiaomi", "converse", "puma", "zara", "jam", "adidas"]
#Recorrer la lista con "for"
for marca in lista_de_marcas:
    #Sacar la longitud de cada elemento a medida que se va recorriendo
    longitud=len(marca)
    #Identificar si las marcas tienen 6 o mas letras
    if longitud>=6 :
        print (marca)
    
```

challenge 7

Solicitar al usuario que ingrese una frase.
Imprimir un mensaje cada vez que se encuentre una letra "o".

```python
# Paso 1: Solicitar el ingreso de la frase y guardar en una variable.

frase = (input("Escriba una frase: "))
#print (frase)

# Paso 2: Recorrer la frase por caracter con for
for letra in frase:
    # Paso 3: Construir una condicional para detectar las letras "o" en la frase
    if letra != "o":
        print (letra)
    elif letra=="o":
        print (f"{letra} --- Se ha encontrado lo solicitado.")

```
