```python
# challenge 5
# Solicitar al usuario ingresar dos números y almacenarlos en dos variables. Luego, imprimir la suma de los dos números.

# 1er paso: Solicitar el ingreso de los datos con INPUT
numero1 = input ("Ingrese el primer numero: ")
print ("PRIMER NUMERO: ", numero1)

numero2 = input ("Ingrese el segundo numero: ")
print ("SEGUNDO NUMERO: ", numero2)

# 2do paso: Convertir los datos ingresados a tipo INT y verificar.
numero1_int = int(numero1)
print("El primer numero es de tipo ", type(numero1_int))
numero2_int = int(numero2)
print ("El segundo numero es de tipo ", type(numero2_int))

#3er paso: Suma de los dos numeros.
suma = numero1_int+numero2_int
print (f"La suma de los numeros da {suma}.")
```
