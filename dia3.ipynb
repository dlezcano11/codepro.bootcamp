```python
palab=input ("Ingrese la palabra a evaluar: ")
letra_a_buscar=input ("Qué letra se buscará en la palabra?: ")

def buscar_letra (l, p):
  #  for letra in p: ------- La funcion "in" ya se puede utilizar para recorrer los caracteres de la palabra.
        if l in p:
            return [l]
        else:
            return l

buscar_letra (letra_a_buscar, palab)
```

challenge 5
Pedir al usuario que ingrese 1 palabra, luego convertir esa palabra a una lista con sus caracteres con list(). Agregar esa lista de caracteres a una lista llamada palabras_ingresadas. Repetir el proceso 3 veces. Cada vez que se agregue una lista de caracteres, imprimir la lista palabras_ingresadas.
> y < 

```python
palabras_ingresadas=[]
for iterador in range(3):
    dato=input (f"Ingrese la palabra numero {iterador+1}: ")
    dato_en_lista=list(dato)
    palabras_ingresadas.append(dato_en_lista)
    print (palabras_ingresadas)
```
